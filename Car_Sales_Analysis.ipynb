### (iv) Creating derived columns

```python
# Car Age
df['Car Age'] = 2023 - df['Year']
```

***Note:-*** To enrich the analysis, a derived feature named Car Age was created from the Year attribute. The age of each vehicle was calculated by subtracting the manufacturing year from 2023. This new attribute helps evaluate how the age of a vehicle influences its market value and resale price.

```python
# Price per Kilometer
df['Price per KM'] = df['Price'] / df['Kilometers']
```

***Note:-*** Helps evaluate the resale value relative to vehicle usage. A higher Price per KM means the vehicle maintains a higher market value despite its usage.
A lower Price per KM means the vehicle's value is relatively low compared to the distance it has been driven.

```python
# Price Category
df['Price Category'] = df['Price'].apply(
    lambda x: 'Low' if x <= 500000
    else 'Medium' if x <= 1000000
    else 'High'
)
```

***Note:-*** The derived Price Category column groups vehicles into Low, Medium, and High price segments. This enables a clearer understanding of the distribution of vehicles across different price ranges and supports comparative analysis of factors such as fuel type, transmission, brand, and vehicle age.

```python
# Owner Number

df['Owner Number'] = df['Owner'].str.extract('(\d+)').astype(int)
```

***Note:-*** This transformation converts categorical ownership information into a numerical format, enabling statistical analysis and easier comparison between vehicles with different ownership histories.

```python
# Usage Category

df['Usage Category'] = pd.cut(
    df['Kilometers'],
    bins=[0, 30000, 70000, float('inf')],
    labels=['Low Usage', 'Medium Usage', 'High Usage']
)
```

***Note:-*** Categorizes vehicles based on distance driven.

```python
# Age Group

df['Age Group'] = pd.cut(
    df['Car Age'],
    bins=[-1, 1, 3, 5],
    labels=['New', 'Moderately Old', 'Old']
)
```

***Note:-*** Groups cars based on age.

```python
# Age Mileage Index

df['Age Mileage Index'] = df['Car Age'] * df['Kilometers']
```

***Note :-*** This metric combines both vehicle age and usage into a single value, helping to estimate the overall wear and tear of a vehicle.  A higher Age Mileage Index indicates an older vehicle with higher usage, suggesting greater wear and depreciation.
A lower Age Mileage Index indicates a newer vehicle with lower usage, suggesting better condition and potentially higher resale value.

```python
# Display main columns and Derived Columns

df[['Year', 'Car Age', 'Kilometers',
    'Usage Category', 'Price', 'Owner Number',
    'Price per KM', 'Price Category', 'Age Mileage Index']].head(20)
```

***Aggregating Data***

Aggregation is used to summarize data and identify trends.

```python
# Average Price by Brand

brand_price = df.groupby('Brand')['Price'].mean().sort_values(ascending=False)

brand_price.round(0).astype(int)
```

***Note :-*** BMW, MG, and Range Rover have the highest average resale prices, indicating strong brand value and demand. In contrast, Ford, Honda, and Mahindra show lower average prices, reflecting higher depreciation and a focus on the budget-friendly segment. Brand reputation significantly influences resale value.

```python
# Average Price by Brand and Owner

brand_price_owner = (
    df[df['Owner'].isin(['1st', '2nd'])]
    .groupby(['Brand', 'Owner'])['Price']
    .mean()
    .round(0)
    .astype(int)
    .unstack()
)

brand_price_owner
```

***Note :-*** Across most brands, 1st Owner vehicles have higher average resale prices than 2nd Owner vehicles. This indicates that buyers are willing to pay a premium for cars with fewer ownership transfers, as they are generally perceived to be in better condition and have a more reliable ownership history.

```python
# Average Price by Fuel Type

fuel_price = (
    df.groupby('Fuel Type')['Price']
      .mean()
      .round(0)
      .astype(int)
      .reset_index()
)

fuel_price.rename(columns={'Price': 'Average Price (₹)'}, inplace=True)

fuel_price
```

```python
# Average Price by Fuel Type and Owner

fuel_price_owner = (
    df[df['Owner'].isin(['1st', '2nd'])]
    .groupby(['Fuel Type', 'Owner'])['Price']
    .mean()
    .round(0)
    .astype(int)
    .unstack()
)

fuel_price_owner
```

***Note :-*** Across all fuel types, 1st-owner vehicles have higher average resale prices than 2nd-owner vehicles. Electric and Hybrid cars command the highest prices, indicating strong market demand and better value retention compared to CNG, Petrol, and Diesel vehicles.

```python
# Number of Cars by Transmission Type

transmission_count = df.groupby('Transmission').size()

transmission_count
```

***Note:-*** Aggregation techniques were used to summarize key metrics across different categories. Grouping vehicles by attributes such as brand, fuel type, and transmission enabled the identification of pricing patterns and market trends. These aggregated insights provide a clearer understanding of the factors influencing used car prices.

* **Visualization Type:** Bar Chart

```python
#To find number of outliers 

Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[
    (df['Price'] < lower_bound) |
    (df['Price'] > upper_bound)
]

outliers
```

Insight :- So only 0.12% of the vehicles are outliers.
Most vehicle prices fall within the expected price range.
Only a small number of vehicles have exceptionally high prices.
These are likely luxury or premium vehicles (BMW, Range Rover, Audi, Jaguar, etc.).
The dataset is relatively clean and does not contain excessive price anomalies.

```python
for fuel in df['Fuel Type'].unique():
    
    fuel_df = df[df['Fuel Type'] == fuel]

    Q1 = fuel_df['Price'].quantile(0.25)
    Q3 = fuel_df['Price'].quantile(0.75)
    IQR = Q3 - Q1

    upper = Q3 + 1.5 * IQR

    outliers = fuel_df[fuel_df['Price'] > upper]

    print(f"{fuel}: {len(outliers)} outliers")
```

Insight :- Petrol (31) and Diesel (29) vehicles account for the majority of price outliers, indicating the presence of several premium models in these segments. CNG has relatively few outliers, while Electric and Hybrid categories show only one outlier each, suggesting a more consistent price distribution within those fuel types.

***Note :-*** Since outliers represent genuine high-value vehicles rather than data errors, removing them may lead to the loss of valuable information about the premium car segment. Therefore, the outliers were retained for analysis.
